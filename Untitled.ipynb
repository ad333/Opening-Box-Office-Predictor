```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.formula.api as sm
```

```python
def backwardElimination(x, sl):
    numVars = len(x[0])
    for i in range(0, numVars):
        regressor_OLS = sm.OLS(y, x).fit()
        maxVar = max(regressor_OLS.pvalues).astype(float)
        if maxVar > sl:
            for j in range(0, numVars - i):
                if (regressor_OLS.pvalues[j].astype(float) == maxVar):
                    x = np.delete(x, j, 1)
    regressor_OLS.summary()
    return x
```

```python
dataset = pd.read_csv('regression_ready_1.csv')
```

```python
X = dataset.iloc[:, 1:].values
y = dataset.iloc[:, 0].values
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
regressor.score(X_test, y_test)
```

```python
import statsmodels.formula.api as sm
X = np.append(arr = np.ones((175, 1)).astype(int), values = X, axis = 1)
X_opt = X[:, :]
SL = 0.1
```

```python
X_Modeled = backwardElimination(X_opt, SL)
X_Modeled = X_Modeled[:, 1:]
```

```python
from sklearn.model_selection import train_test_split
X_opt_train, X_opt_test, y_opt_train, y_opt_test = train_test_split(X_Modeled, y, test_size = 0.1, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
opt_regressor = LinearRegression()
opt_regressor.fit(X_opt_train, y_opt_train)
```

```python
opt_y_pred = opt_regressor.predict(X_opt_test)
opt_regressor.score(X_opt_test, y_opt_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_opt_test,opt_y_pred)
```
