```python
import pandas as pd
import numpy as np
```

```python
bo_data = "before_numeric_conversion.csv"
```

```python
df = pd.read_csv(bo_data)
```

```python
df.head()
```

```python
df.count()
```

```python
df.nunique()
```

```python
avg_per_studio = df.groupby("Studio")["Opening"].mean()
avg_per_studio.astype('int64')
```

```python
avg_per_studio_df = pd.DataFrame(avg_per_studio)
avg_per_studio_df=round(avg_per_studio_df)
avg_per_studio_df.columns = ['ave_opening_by_studio']
avg_per_studio_df.reset_index()
```

```python
combine_df = df.join(avg_per_studio_df, on="Studio")
combine_df.head()
```

```python
# Determine maximum and minimum Opening values to create bins
print("max:", avg_per_studio.max())
print("min:", avg_per_studio.min())
```

```python
# Create bins and assign group labels
bins = [0, 50000000, 65000000, 90000000, 105000000, 120000000, 135000000, 150000000]
group_labels = ["1","2","3","4","5","6","7"]

# Slice the data and place into bins and put as a new column
combine_df["Studio_value"] = pd.cut(combine_df["ave_opening_by_studio"], bins, labels = group_labels)
combine_df.head()
```

```python
combine_df.to_csv('after_numeric_conversion.csv')
```
