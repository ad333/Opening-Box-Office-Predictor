```python
import pandas as pd
import requests as req
import json
```

```python
# Read box_office_data.csv which includes ticket price
weekend_boxfinal = pd.read_csv('box_office_data.csv')
```

```python
weekend_boxfinal.head(12)
```

```python
# Drop the extra index column
weekend_boxfinal = weekend_boxfinal.drop(weekend_boxfinal.columns[[0]], axis=1)
```

```python
# Add variables
weekend_boxfinal['Actor'] = ''
weekend_boxfinal['Genre'] = ''
weekend_boxfinal['Writer'] = ''
weekend_boxfinal['Ratings'] = ''
```

```python
# Call OMDB API to add actors, genres, writer, ratings
omdb = 'http://www.omdbapi.com/?apikey=7b298284&t='
```

```python
# Loop through titles and create title format accepted by OMDB
# Add .head after weekend_box to test/limit API calls - default number is 5
for index, row in weekend_boxfinal.iterrows():
    t = str(row['Title'])
    t = t.replace(":","")
    t = t.replace("'","")
    t = t.replace(' ', '+')
    search = omdb + t
    print (search)
    try:
        results = req.get(search).json()
        if results['Response'] == 'False':
            print("Couldn't find title: ", t)
        else:
            weekend_boxfinal.set_value(index, 'Actor', results['Actors'])
            weekend_boxfinal.set_value(index, 'Genre', results['Genre'])
            weekend_boxfinal.set_value(index, 'Writer', results['Writer'])
            weekend_boxfinal.set_value(index, 'Ratings', results['Ratings'][1])
    except Exception as e:
        print(str(e))
        continue
```

```python
for index, row in weekend_boxfinal.iterrows():
    try:
        rating = row['Ratings']
        rating = rating['Value']
        rating = rating.replace('%','')
        rating = float(rating)
        rating = rating/100
    except:
        continue    
    weekend_boxfinal.set_value(index, 'Ratings', rating)
```

```python
weekend_boxfinal.rename(columns={'month': 'Month', 'year': 'Year', 'ticket_price': 'Ticket_Price'}, inplace=True)
```
